As there are many NULL values present in the given dataset. We need to replace those values by mean(in case of numerical data) or mode(in case of categorical data). Here, we need to work on "First Dose Administered" and "Second Dose Administered". Both of them are float, hence we will replace the Nan Values by mean(average).

#### For First Dose Administered

```python
# Average of First Dose Administered
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
# Replacing First Dose Administered
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

#### For Second Dose Administered

```python
# Average of Second Dose Administered
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

```python
# Replacing Second Dose Administered
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import squarify  # For treemap
import plotly.express as px
import numpy as np

# Load the data
data = pd.read_csv('/content/covid_vaccine_statewise.csv')
india_data = data[data['State'] == 'India']
dates = pd.to_datetime(india_data['Updated On'], format='%d/%m/%Y')
```

```python
# 1. Line Chart
plt.figure(figsize=(10,6))
plt.plot(dates, india_data['Total Doses Administered'], label='Total Doses Administered', marker='o', color='b')
plt.title('Total COVID-19 Vaccine Doses Administered Over Time (India)')
plt.xlabel('Date')
plt.ylabel('Total Doses Administered')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# 2. Bar Chart
plt.figure(figsize=(10,6))
plt.bar(dates, india_data['First Dose Administered'], label='First Dose', color='g', alpha=0.7)
plt.bar(dates, india_data['Second Dose Administered'], label='Second Dose', color='r', alpha=0.7)
plt.title('First vs Second Dose Administered Over Time (India)')
plt.xlabel('Date')
plt.ylabel('Number of Doses Administered')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# 3. Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(dates, india_data['Total Doses Administered'], label='Total Doses', color='purple')
plt.title('Scatter Plot of Doses Administered Over Time (India)')
plt.xlabel('Date')
plt.ylabel('Total Doses Administered')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#4.  pie chart
gender_distribution = india_data[['Male(Individuals Vaccinated)', 'Female(Individuals Vaccinated)', 'Transgender(Individuals Vaccinated)']].iloc[-1]

# Check for NaN values and handle them
print("Gender Distribution Data Before Dropping NaNs:")
print(gender_distribution)

# Drop any NaN values if found
gender_distribution = gender_distribution.dropna()

# Check if the data is empty after dropping NaN values
if gender_distribution.empty:
    print("The gender distribution data is empty after dropping NaNs. Cannot generate pie chart.")
else:
    # Assigning custom labels for the pie chart
    gender_distribution.index = ['Male', 'Female', 'Transgender']

    # Plotting the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink', 'purple'])
    plt.title('Gender Distribution of Vaccinated Individuals (India)')
    plt.show()
```

```python
# 5. Column Chart
plt.figure(figsize=(10,6))
plt.bar(dates, india_data['Total Individuals Vaccinated'], color='orange')
plt.title('Total Individuals Vaccinated Over Time (India)')
plt.xlabel('Date')
plt.ylabel('Total Individuals Vaccinated')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Treemap Chart
plt.figure(figsize=(10,6))
states = data['State'].unique()
statewise_total_vaccination = data.groupby('State')['Total Doses Administered'].sum().dropna()
squarify.plot(sizes=statewise_total_vaccination, label=statewise_total_vaccination.index, alpha=0.8)
plt.title('Treemap of Total Vaccinations by State')
plt.axis('off')
plt.show()
```

```python
# 7. Heatmap
plt.figure(figsize=(10,6))
heatmap_data = data.pivot_table(values='Total Doses Administered', index='State', columns='Updated On', aggfunc='sum')
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Heatmap of Doses Administered by State and Date')
plt.show()
```

```python
# 8. Pareto Chart
plt.figure(figsize=(10,6))
sorted_data = india_data['Total Doses Administered'].sort_values(ascending=False).cumsum()
plt.plot(sorted_data, label="Cumulative Doses")
plt.bar(range(len(sorted_data)), india_data['Total Doses Administered'], color="C0", alpha=0.7)
plt.title('Pareto Chart: Cumulative Doses Administered (India)')
plt.xlabel('Date')
plt.ylabel('Doses Administered')
plt.show()
```

```python
# 9. Donut Chart (like Pie but with a hole in the middle)
plt.figure(figsize=(8,8))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink', 'purple'], wedgeprops={'width':0.4})
plt.title('Donut Chart of Gender Distribution (India)')
plt.show()
```

```python
# 10. Bubble Chart
plt.figure(figsize=(10,6))
plt.scatter(dates, india_data['Total Doses Administered'], s=india_data['First Dose Administered'] / 10, alpha=0.5, label="Total Doses")
plt.title('Bubble Chart of Doses Administered (India)')
plt.xlabel('Date')
plt.ylabel('Total Doses Administered')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python

# 11. Histogram
plt.figure(figsize=(10,6))
plt.hist(india_data['Total Doses Administered'], bins=10, color='c', edgecolor='k')
plt.title('Histogram of Total Doses Administered (India)')
plt.xlabel('Total Doses Administered')
plt.ylabel('Frequency')
plt.show()
```

```python
# 12. Area Chart
plt.figure(figsize=(10,6))
plt.fill_between(dates, india_data['Total Doses Administered'], color='skyblue', alpha=0.4)
plt.title('Area Chart of Doses Administered Over Time (India)')
plt.xlabel('Date')
plt.ylabel('Total Doses Administered')
plt.xticks(rotation=45)
plt.show()
```
